```python
from PIL import Image
import numpy as np

# Load an image from a file
image = Image.open('NumPy.webp')

# Convert the image to a NumPy array
image_array = np.array(image)

# Save the array to a .npy file
np.save('image.npy', image_array)
import matplotlib.pyplot as plt
with open('image.npy','rb') as f:
    logo_rgb_array = np.load(f)
    plt.imshow(logo_rgb_array)
    plt.show
```

```python
red_array,green_array,blue_array = np.split(logo_rgb_array,3,axis=2)
plt.imshow(red_array)
plt.show
```

```python
# stacking 2d arrays
red_array = np.zeros((1001,1001)).astype(np.int32)
green_array = np.random.randint(0, 256, (1001, 1001)).astype(np.int32)  
blue_array = np.random.randint(0, 256, (1001, 1001)).astype(np.int32) 
stacked_rgb = np.stack([red_array,green_array,blue_array],axis = 2)
plt.imshow(stacked_rgb)
plt.show()
```
