```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load an image from a file
image = Image.open('NumPy.webp')

# Convert the image to a NumPy array
image_array = np.array(image)

# Save the array to a .npy file
np.save('image.npy', image_array)
```

```python
with open('image.npy','rb') as f:
    logo_rgb_array = np.load(f)
    plt.imshow(logo_rgb_array)
    plt.show()
```

```python
flipped_logo = np.flip(logo_rgb_array,axis=(0,1))
plt.imshow(flipped_logo)
plt.show() 
```

```python
array = np.array([
    [0,2,1],
    [6,4,0],
    [2,3,1],
    [0,0,1],
    [2,2,0]])
np.flip(array)
```

```python
transposed = np.transpose(logo_rgb_array,axes=(1,0,2))
plt.imshow(transposed)
plt.show()
```
