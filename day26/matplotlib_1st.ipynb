```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
plt.show()
```

```python
seattle_weather = {
    "months": ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"],
    "avg_temps": [45, 47, 50, 55, 60, 65, 70, 71, 65, 57, 50, 45]
}
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["months"],seattle_weather["avg_temps"])
plt.show()
```

```python
austin_weather = {
    "months": ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"],
    "avg_temps": [55, 58, 64, 71, 78, 85, 89, 88, 82, 72, 62, 56]
}
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["months"],seattle_weather["avg_temps"],marker='o',linestyle='none',color='g')
ax.plot(austin_weather["months"],austin_weather["avg_temps"],marker='v',linestyle='--',color='b')
ax.set_xlabel("time(months)")
ax.set_ylabel("average temperature")
ax.set_title("austin and seattle average temp..")
plt.show()
```
